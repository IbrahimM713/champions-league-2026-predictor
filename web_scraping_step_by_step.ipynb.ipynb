```python
# Selecting which years to pull data from
years = list(range(2026, 2023, -1))
years 
```

```python
all_matches = []
standings_url = "https://fbref.com/en/comps/8/Champions-League-Stats"
```

```python
for year in years:
    # Doing the same steps as before to scrape the links for each team
    # Pulling the HTML data from the link
    data = scraper.get(standings_url)
    # Getting the HTML only from the standings table
    soup = BeautifulSoup(data.text)
    standings_table = soup.select("table.stats_table")[0]

    # Finding all of the a tags within the table, which contains the link to more data for each team
    links = standings_table.find_all("a")
    # Extracting the href properties from the links
    links = [link.get("href") for link in links]
    # Filtering the links to only include those that contain "/squads/", which is the pattern for team data pages
    links = [l for l in links if l and "/squads/" in l]
    # Constructing the full URLs for each team
    team_urls = [f"https://fbref.com{l}" for l in links]

    # Scraping the "previous season" button" by findings its href
    previous_season= soup.select("a.prev")[0].get("href")
    standings_url = f"https://fbref.com{previous_season}"

    for team_url in team_urls:
        # Changing the Team name from the url to a readable format (From "Manchester-City-Stats" to "Manchester City")
        team_name = team_url.split("/")[-1].replace("-", " ").replace("_Stats", "")

        data = scraper.get(team_url)
        matches = pd.read_html(StringIO(data.text), match="Scores & Fixtures")

        # Pulling all the data related to shooting
        # Same steps as before to get link 
        soup = BeautifulSoup(data.text)
        page_links = soup.find_all("a")
        page_links = [link.get("href") for link in page_links]
        page_links = [l for l in page_links if l and "all_comps/shooting/" in l]

        data = scraper.get(f"https://fbref.com{page_links[0]}")
        shooting = pd.read_html(StringIO(data.text), match="Shooting")[0]
        # Dropping the first row of headers (there is initially two rows of eaders in "shooting")
        shooting.columns = shooting.columns.droplevel()

        # Trying to merge the team match and shooting stats. If there is not shooting stats, it skips those values and states what team and year it skipped
        try:
            team_data = matches[0].merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
            team_data["Team"] = team_name
            team_data["Season"] = f"{year-1}-{str(year)[-2:]}"
            all_matches.append(team_data)
        except ValueError:
            print(f"Skipping {team_name} for season {year-1}-{str(year)[-2:]} due to merge error.")
            continue

            # Adding the team name and season year to the dataframe to ensure we kn ow which team the result is for
        team_data["Season"] = year
        team_data["Team"] = team_name

        all_matches.append(team_data)
        time.sleep(1)  # Be polite and avoid overwhelming the server

```

```python
match_df = pd.concat(all_matches)
```
