# Beginning of Creation of the Model

This begins the code where we develop the Random Forest Model. Random Forest is used because it can pick up non-linear relationships and handle both team and opponent features effectively.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import precision_score
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1, class_weight="balanced")

# Single function necessary to make predictions

def make_predictions(data, predictors):
    train = data[data["Date"] < '2025-01-01']
    test = data[data["Date"] > '2025-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined_df = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    precision = precision_score(test["target"], preds, average="macro")
    accuracy = accuracy_score(test["target"], preds)
    
    return combined_df, precision, accuracy
```

```python
# Features for the model - using columns that exist in both datasets
predictors = [
 'team_elo', 'opp_elo', 'elo_diff',  'prev_team_elo',
 'prev_opp_elo', 'form_elo', 'opp_form_elo', 'form_elo_diff', 'avg_xGA',
 'avg_Sh_shooting', 'avg_Dist_shooting', 'avg_Def 3rd_poss',
 'avg_Mid 3rd_poss', 'avg_Succ%_poss', 'opp_avg_Succ%_poss', 'opp_avg_Mid 3rd_poss',
    'opp_avg_Def 3rd_poss', 'avg_Att 3rd_poss',
]

```

```python
combined_df, precision, accuracy = make_predictions(matches_past, predictors)
```

```python
# SIMPLE BIDIRECTIONAL PREDICTIONS
# Get predictions from both perspectives and average them

def get_simple_bidirectional_predictions(df, predictors, model):
    """Simple function to get balanced predictions from both team perspectives"""
    
    # Get original predictions
    original_probs = model.predict_proba(df[predictors])
    
    # Create reversed data by swapping team/opponent features
    reversed_df = df.copy()
    
    # Swap the key features
    reversed_df['team_elo'], reversed_df['opp_elo'] = df['opp_elo'], df['team_elo']
    reversed_df['prev_team_elo'], reversed_df['prev_opp_elo'] = df['prev_opp_elo'], df['prev_team_elo'] 
    reversed_df['form_elo'], reversed_df['opp_form_elo'] = df['opp_form_elo'], df['form_elo']
    reversed_df['elo_diff'] = -df['elo_diff']
    reversed_df['form_elo_diff'] = -df['form_elo_diff']
    
    # Swap rolling averages
    for col in predictors:
        if col.startswith('avg_') and not col.startswith('opp_'):
            opp_col = f"opp_{col}"
            if opp_col in predictors:
                reversed_df[col], reversed_df[opp_col] = df[opp_col], df[col]
    
    # Get reversed predictions and flip them
    reversed_probs = model.predict_proba(reversed_df[predictors])
    flipped_probs = reversed_probs[:, [2, 1, 0]]  # Flip [loss, draw, win] to [win, draw, loss]
    
    # Average the probabilities
    final_probs = (original_probs + flipped_probs) / 2
    final_preds = final_probs.argmax(axis=1)
    
    return final_probs, final_preds

print("Created simple bidirectional prediction system!")
```

```python
# Create a unique match identifier by sorting team names (with normalization)
def normalize_team_name(name):
    """Normalize team names to handle encoding differences"""
    # Replace common character variations
    normalized = name.replace('ı', 'i').replace('İ', 'I')
    normalized = normalized.replace('ö', 'o').replace('ü', 'u')
    normalized = normalized.replace('ç', 'c').replace('ş', 's')
    normalized = normalized.strip()
    return normalized

def create_match_id(row):
    team1 = normalize_team_name(row['Team'])
    team2 = normalize_team_name(row['Opponent'])
    teams = sorted([team1, team2])
    return f"{row['Date']}_{teams[0]}_vs_{teams[1]}"

# Apply simple bidirectional predictions to Champions League matches
ucl_future = matches_future[matches_future["Comp"].str.contains("Champions Lg", case=False)].copy()

# Get balanced predictions
final_probs, final_preds = get_simple_bidirectional_predictions(ucl_future, predictors, rf)

# Add results to dataframe
ucl_future["balanced_pred"] = final_preds
ucl_future["balanced_prob_loss"] = final_probs[:, 0]
ucl_future["balanced_prob_draw"] = final_probs[:, 1] 
ucl_future["balanced_prob_win"] = final_probs[:, 2]

# Map predictions to text
prediction_map = {0: 'Loss', 1: 'Draw', 2: 'Win'}
ucl_future["prediction"] = ucl_future["balanced_pred"].map(prediction_map)

# IMPROVED DUPLICATE REMOVAL - Remove ALL duplicates properly
ucl_future['match_id'] = ucl_future.apply(create_match_id, axis=1)

# SIMPLE FIX: Just drop duplicates by match_id and keep first occurrence
print(f"Before duplicate removal: {len(ucl_future)} matches")
ucl_unique = ucl_future.drop_duplicates(subset='match_id', keep='first').reset_index(drop=True)
print(f"After duplicate removal: {len(ucl_unique)} matches")

# For any remaining potential duplicates, let's manually check and fix
# Group by match_id and average probabilities if there are still any duplicates
grouped = ucl_unique.groupby('match_id')
final_matches = []

for match_id, group in grouped:
    if len(group) == 1:
        # Single match - keep as is
        final_matches.append(group.iloc[0].to_dict())
    else:
        # Multiple matches with same ID - shouldn't happen now, but just in case
        print(f"Warning: Still found {len(group)} matches with ID {match_id}")
        # Take the first one
        final_matches.append(group.iloc[0].to_dict())

# Convert back to dataframe
ucl_unique = pd.DataFrame(final_matches)

# Display results
result_cols = ['Date', 'Team', 'Opponent', 'prediction', 'balanced_prob_win', 'balanced_prob_draw', 'balanced_prob_loss']
print("=== CHAMPIONS LEAGUE UNIQUE MATCH PREDICTIONS ===")
print(ucl_unique[result_cols].round(3).to_string(index=False))

print(f"\nPrediction Summary:")
print(ucl_unique['prediction'].value_counts())
print(f"\nTotal unique matches: {len(ucl_unique)} (reduced from {len(ucl_future)} entries)")
```

```python
# CREATE CHAMPIONS LEAGUE TABLE FROM PREDICTIONS
print("=== CHAMPIONS LEAGUE TABLE FROM PREDICTIONS ===")

# Initialize team stats dictionary
team_stats = {}

# Process each predicted match
for idx, match in ucl_unique.iterrows():
    team = match['Team']
    opponent = match['Opponent']
    prediction = match['prediction']
    
    # Initialize teams if not exists
    if team not in team_stats:
        team_stats[team] = {'wins': 0, 'draws': 0, 'losses': 0, 'points': 0, 'games': 0}
    if opponent not in team_stats:
        team_stats[opponent] = {'wins': 0, 'draws': 0, 'losses': 0, 'points': 0, 'games': 0}
    
    # Update stats based on prediction
    if prediction == 'Win':  # Team wins
        team_stats[team]['wins'] += 1
        team_stats[team]['points'] += 3
        team_stats[opponent]['losses'] += 1
        team_stats[opponent]['points'] += 0
    elif prediction == 'Draw':  # Draw
        team_stats[team]['draws'] += 1
        team_stats[team]['points'] += 1
        team_stats[opponent]['draws'] += 1
        team_stats[opponent]['points'] += 1
    else:  # Team loses (opponent wins)
        team_stats[team]['losses'] += 1
        team_stats[team]['points'] += 0
        team_stats[opponent]['wins'] += 1
        team_stats[opponent]['points'] += 3
    
    # Update games played
    team_stats[team]['games'] += 1
    team_stats[opponent]['games'] += 1

# Convert to DataFrame and sort by points (then by wins as tiebreaker)
table_data = []
for team, stats in team_stats.items():
    table_data.append({
        'Team': team,
        'Games': stats['games'],
        'Wins': stats['wins'],
        'Draws': stats['draws'],
        'Losses': stats['losses'],
        'Points': stats['points']
    })

# Create DataFrame and sort by points (descending), then wins (descending)
cl_table = pd.DataFrame(table_data)
cl_table = cl_table.sort_values(['Points', 'Wins'], ascending=[False, False]).reset_index(drop=True)

# Add position column
cl_table['Position'] = range(1, len(cl_table) + 1)

# Reorder columns for better display
cl_table = cl_table[['Position', 'Team', 'Games', 'Wins', 'Draws', 'Losses', 'Points']]

print("\nCHAMPIONS LEAGUE TABLE (Based on Predictions):")
print("=" * 70)
print(cl_table.to_string(index=False))

print(f"\n=== QUALIFICATION SCENARIOS ===")
print(f"🏆 Automatic Round of 16 (Top 8): {', '.join(cl_table.head(8)['Team'].tolist())}")
print(f"🎯 Playoff Round (9th-24th): {', '.join(cl_table.iloc[8:24]['Team'].tolist())}")
if len(cl_table) > 24:
    print(f"❌ Eliminated (25th+): {', '.join(cl_table.tail(len(cl_table)-24)['Team'].tolist())}")

print(f"\n=== TABLE STATISTICS ===")
print(f"Total teams: {len(cl_table)}")
print(f"Total matches processed: {len(ucl_unique)}")
print(f"Average points per team: {cl_table['Points'].mean():.1f}")
print(f"Most wins: {cl_table.iloc[0]['Team']} ({cl_table.iloc[0]['Wins']} wins)")
print(f"Highest scoring: {cl_table.iloc[0]['Team']} ({cl_table.iloc[0]['Points']} points)")
```

```python
# DEBUG: Check for duplicate matches and game counts
print("=== DEBUGGING GAME COUNTS ===")

# Check for teams with unexpected game counts
print("Teams with more than 8 games:")
for team, stats in team_stats.items():
    if stats['games'] > 8:
        print(f"{team}: {stats['games']} games")

# Check if there are any duplicate match IDs that weren't caught
print(f"\nUnique match IDs: {ucl_unique['match_id'].nunique()}")
print(f"Total rows: {len(ucl_unique)}")

# Show matches for a team with 9 games to investigate
problematic_teams = [team for team, stats in team_stats.items() if stats['games'] > 8]
if problematic_teams:
    problem_team = problematic_teams[0]
    print(f"\nMatches for {problem_team}:")
    team_matches = ucl_unique[(ucl_unique['Team'] == problem_team) | (ucl_unique['Opponent'] == problem_team)]
    print(team_matches[['Date', 'Team', 'Opponent', 'match_id']].to_string(index=False))
    
    # Check for any duplicate match_ids
    duplicate_ids = team_matches['match_id'].value_counts()
    duplicates = duplicate_ids[duplicate_ids > 1]
    if len(duplicates) > 0:
        print(f"\nDuplicate match IDs found:")
        print(duplicates)

# Check if any matches are counted twice
print(f"\nMatch ID value counts (should all be 1):")
match_counts = ucl_unique['match_id'].value_counts()
duplicates = match_counts[match_counts > 1]
if len(duplicates) > 0:
    print("Found duplicates:")
    print(duplicates)
else:
    print("No duplicates found in match_id")
```

```python
# CHECK FOR QAIRAT ALMATY VARIATIONS AND FIX THEM
print("=== CHECKING FOR QAIRAT ALMATY VARIATIONS ===")

# Let's check the raw data before normalization
print("Checking raw team names...")
all_teams = set(ucl_unique['Team'].tolist() + ucl_unique['Opponent'].tolist())
qairat_variations = [team for team in all_teams if 'airat' in team.lower()]
print(f"All team name variations containing 'airat': {qairat_variations}")

# Apply normalization to actual team names in the dataframe
print("\nApplying team name normalization to dataframe...")
ucl_unique['Team'] = ucl_unique['Team'].apply(normalize_team_name)
ucl_unique['Opponent'] = ucl_unique['Opponent'].apply(normalize_team_name)

# Check after normalization
all_teams_normalized = set(ucl_unique['Team'].tolist() + ucl_unique['Opponent'].tolist())
qairat_after = [team for team in all_teams_normalized if 'airat' in team.lower()]
print(f"After normalization: {qairat_after}")

# Recreate match IDs with normalized names
ucl_unique['match_id'] = ucl_unique.apply(create_match_id, axis=1)

# Remove any duplicates created by normalization
print(f"Before final deduplication: {len(ucl_unique)} matches")
ucl_unique_final = ucl_unique.drop_duplicates(subset='match_id', keep='first').reset_index(drop=True)
print(f"After final deduplication: {len(ucl_unique_final)} matches")

# Update the global ucl_unique variable
ucl_unique = ucl_unique_final

print("✅ Team names normalized and duplicates removed!")

# Show Qairat matches after normalization
qairat_matches = ucl_unique[(ucl_unique['Team'].str.contains('airat', case=False, na=False)) | 
                           (ucl_unique['Opponent'].str.contains('airat', case=False, na=False))]
print(f"\nFinal Qairat Almaty matches:")
if len(qairat_matches) > 0:
    print(qairat_matches[['Date', 'Team', 'Opponent']].to_string(index=False))
else:
    print("None found")
```
